```python
# Import dependencies
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
# Set up the executable path
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = "http://quotes.toscrape.com/"
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
# Scrape the Title
title = soup.find("h2").text
title
```

```python
# Scrape the top ten tags
tag_box = soup.find("div", class_="tags-box")

tags = tag_box.find_all("a", class_="tag")

for tag in tags:
    word = tag.text
    print(word)
```

```python
# Create a for loop to collect each quote, go to the next, collect each quote, etc
for x in range(1, 6):
    html = browser.html
    soup = BeautifulSoup(html, "html.parser")
    quotes = soup.find_all("span", class_="text")
    for quote in quotes:
        print("page", x, "----------")
        print(quote.text)
    browser.click_link_by_partial_text("Next")
```

```python
# 10.3.2 SKILL DRILL
url = "http://books.toscrape.com/"
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, "html.parser")
books = soup.find_all("ul", class_="nav")

for book in books:
    word = book.text
    print(word)
```
